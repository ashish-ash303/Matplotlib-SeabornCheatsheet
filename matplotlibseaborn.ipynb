```python
# Plotting Multiple Lines 
years=range(2010,2020)
apples=[0.222,0.993,0.555,0.666,0.777,0.455,0.553,0.444,0.333,0.22]
oranges=[0.212,0.888,0.992,0.444,0.656,0.776,0.345,0.556,0.877,0.321]
```

```python
plt.plot(years,apples);
plt.plot(years,oranges);
plt.xlabel("Year")
plt.ylabel("Yeild (tones Per hectare )");
```

## First Conclusion:-
This shows How the Yeild of Apples and Oranges Were  in the Past 10 Years 
Conclusion:- We Conclude that the demand of the Apples ans Oranges Were very Varying in Past 10 years. For some years  demand  Were riseing Continously while for some Years demand Sees a great dip this is Probably due to less Production of fruits.

```python
# Chart Time and legend
```

```python
plt.plot(years,apples);
plt.plot(years,oranges);
plt.xlabel("Year")
plt.ylabel("Yeild (tones Per hectare )");

plt.title("Crop Yeilds in India")
plt.legend(["Apples","Orange"]);
```

```python
# Line Marker Marker are like different  Point on  line 

plt.plot(years,apples,marker='o')
plt.plot(years,oranges,marker='x')

# plt.plot(years,oranges);
plt.xlabel("Year")
plt.ylabel("Yeild (tones Per hectare )");

plt.title("Crop Yeilds in India")
plt.legend(["Apples","Orange"]);
```

```python
# Styling lines & markers plt.plot Support differnt type of arguement like colour ,linestyle=ls,linewidth=lw,
# ms:- marker size=ms, mew=marker width,mec=marker color ,alpha=opacity


plt.plot(years,apples,marker='s',c='b',ls='-',lw=2,ms=8,mew=2,mec='navy');
plt.plot(years,oranges,marker='o',c='r',ls='--',lw=3,ms=10,mew=2,alpha=.5);
```

```python
#fmt arguement is a shorthand which contain line style ,marker& line color fmt=[marker][line][color] ex:-s--o s is for marker,
# --is for line b is for color
```

```python

plt.plot(years,apples,'s--b')
plt.plot(years,oranges,'o--r')

# plt.plot(years,oranges);
plt.xlabel("Year")
plt.ylabel("Yeild (tones Per hectare )");

plt.title("Crop Yeilds in India")
plt.legend(["Apples","Orange"]);
```

```python
# if we not want line we can give arguement like this so
plt.plot(years,apples,'sb')
plt.plot(years,oranges,'or')

# plt.plot(years,oranges);
plt.xlabel("Year")
plt.ylabel("Yeild (tones Per hectare )");

plt.title("Crop Yeilds in India")
plt.legend(["Apples","Orange"]);
```

```python
flower_df=sns.load_dataset("iris")
# dataset of scater
```

```python
plt.plot(flower_df.sepal_length,flower_df.sepal_width);
```

```python
## We get this Scattered Value Bcz Values are too Close SO We Use Scatterd Plot
```

```python
sns.scatterplot(flower_df.sepal_length,flower_df.sepal_width);
```

```python
sns.scatterplot(flower_df.sepal_length,flower_df.sepal_width,hue=flower_df.species,s=150);
# Adding Hue gives us Some Clarity about the data like their are 3 species & each Species has How Lenght &breath
```

```python
plt.figure(figsize=(16,8))
plt.title('Sepal Dimension')
sns.scatterplot(flower_df.sepal_length,flower_df.sepal_width,hue=flower_df.species,s=150);
```

```python
plt.figure(figsize=(16,8))
plt.title('Sepal Dimension')
sns.scatterplot(flower_df.sepal_length,flower_df.sepal_width,hue=flower_df.species,s=150,data=flower_df);
```

```python
setosa_df=flower_df[flower_df.species=='setosa']
versicolor_df=flower_df[flower_df.species=='versicolor']
virginica_df=flower_df[flower_df.species=='virginica']
```

```python
setosa_df
```

```python
plt.hist(setosa_df.sepal_width,alpha=0.4,bins=np.arange(2,5,0.25));
plt.hist(versicolor_df.sepal_width,alpha=0.4,bins=np.arange(2,5,0.25));
plt.legend(['setosa','versicolor']);
```

```python
plt.title('Distribution Of Sepal Width')
plt.hist([setosa_df.sepal_width,versicolor_df.sepal_width,virginica_df.sepal_width]
           ,bins=np.arange(2,5,0.25),stacked=True);

plt.legend(['setosa','versicolor','virginica']);
```
